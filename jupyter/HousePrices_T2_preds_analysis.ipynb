### Less Than 300K

```python
preds_lt300k = preds.loc[preds.SalePrice <= 300000].sort_values(by='SalePrice', ascending=False)[:30]

fig = px.scatter(preds_lt300k, x="TOTAL_FLR_SQFT", y="SalePrice", 
                 size="TOTAL_FLR_SQFT_PER_LOT_AREA", color="HouseStyle",
           hover_name="Street", 
                 log_x=False, 
                 size_max=60)
fig.show()
```

```python
preds_lt300k = preds.loc[preds.SalePrice >= 300000].sort_values(by='SalePrice', ascending=False)[:30]

fig = px.scatter(preds_lt300k, x="TOTAL_FLR_SQFT", y="SalePrice", 
                 size="OverallQual", color="HouseStyle",
           hover_name="Street", 
                 log_x=False, 
                 size_max=60)
fig.show()
```

```python
preds_lt300k = preds.loc[preds.SalePrice >= 300000].sort_values(by='SalePrice', ascending=False)[:30]

fig = px.scatter(preds_lt300k, x="TOTAL_FLR_SQFT", y="SalePrice", 
                 size="LotArea", color="HouseStyle",
           hover_name="Street", 
                 log_x=False, 
                 size_max=60)
fig.show()
```
